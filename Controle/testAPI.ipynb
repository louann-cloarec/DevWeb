```python
import requests
import json
from PIL import Image
from io import BytesIO
import requests
from urllib.parse import urlencode

class diabeteSDK:
    def __init__(self, api_key, base_url="http://127.0.0.1:5000"):
        self.api_key = api_key
        self.base_url = base_url.rstrip("/")

    def predict(self, *args, **kwargs):
        response = requests.post(
            f"{self.base_url}/api/predict",
            headers={
                "X-API-Key": self.api_key,
                "Content-Type": "application/json",
                "Body": json.dumps(kwargs),
            },
        )
        data = response.json()
        if response.status_code != 200:
            raise Exception(f"Erreur API ({response.status_code}) : {data}")
        return data

    def waterfall(self, *args, **kwargs):
        response = requests.get(
            f"{self.base_url}/api/waterfall?{urlencode(kwargs)}",
            headers={"X-API-Key": self.api_key},
        )
        if response.status_code != 200:
            data = response.json()
            raise Exception(f"Erreur API ({response.status_code}), : {data}")
        img = Image.open(BytesIO(response.content))
        display(img)

diabete = diabeteSDK(api_key="njNq4wq1wQtXcispXf3UxcqwLuMQ9dJFS5HTTruskuE")
```

```python
diabete.predict(Pregnancies=6.0, Glucose=148.0, BloodPressure=72.0, SkinThickness=35.0, Insulin=0.0, BMI=33.6, Age=50.0)
```
